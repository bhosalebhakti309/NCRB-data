```python
#ax=df.groupby(df['child victim below 6yr'].sum(),df['6-12'].sum(),df['12-16'].sum(),df['16 -18'].sum()).plot(kind='bar')
#for bars in ax.containers:
 # ax.bar_label(bars)
#plt.xticks(rotation=90)
#plt.title('age wise affected group ')
import matplotlib.pyplot as plt

# Sum  values for each age group
age_groups = {
    'Below 6': df['child victim below 6yr'].sum(),
    '6-12': df['6-12'].sum(),
    '12-16': df['12-16'].sum(),
    '16-18': df['16 -18'].sum()
}

# Create the bar plot
fig, ax = plt.subplots()
bars = ax.bar(age_groups.keys(), age_groups.values())

# Add labels on top of each bar
ax.bar_label(bars)

# Formatting
plt.xticks(rotation=0)
plt.title('Age-wise Affected Child Victim Count')
plt.xlabel('Age Group')
plt.ylabel('Number of Victims')
plt.show()
```
