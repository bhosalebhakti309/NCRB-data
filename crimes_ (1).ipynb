```python
#ax=df.groupby(df['child victim below 6yr'].sum(),df['6-12'].sum(),df['12-16'].sum(),df['16 -18'].sum()).plot(kind='bar')
#for bars in ax.containers:
 # ax.bar_label(bars)
#plt.xticks(rotation=90)
#plt.title('age wise affected group ')
import matplotlib.pyplot as plt

# Sum the values for each age group
age_groups = {
    'Below 6': df['child victim below 6yr'].sum(),
    '6-12': df['6-12'].sum(),
    '12-16': df['12-16'].sum(),
    '16-18': df['16 -18'].sum()
}

# Create the bar plot
fig, ax = plt.subplots()
bars = ax.bar(age_groups.keys(), age_groups.values())

# Add labels on top of each bar
ax.bar_label(bars)

# Formatting
plt.xticks(rotation=0)
plt.title('Age-wise Affected Child Victim Count')
plt.xlabel('Age Group')
plt.ylabel('Number of Victims')
plt.tight_layout()
plt.show()
```

```python
# Sum each age group per state
age_data = df.groupby('State/UT')[['child victim below 6yr', '6-12', '12-16', '16 -18']].sum()

# Plot as a stacked bar chart
ax = age_data.plot(kind='bar', stacked=True, figsize=(12, 6))

# Add labels
for bars in ax.containers:
    ax.bar_label(bars, label_type='center', fontsize=7)

plt.title('Age-wise Minor Victims by State')
plt.xlabel('State/UT')
plt.ylabel('Number of Victims')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
